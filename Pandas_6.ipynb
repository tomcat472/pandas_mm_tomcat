```python
# unique
# unique ဖြစ်တဲ့တန်ဖိုးတွေကိုရှာခြင်း
df.color.unique()
# df['color'].unique()
# ထွက်လာတဲ့ output တန်ဖိုးတွေကိုကြည့်ရင် unique ဖြစ်တာတွေကိုတွေ့ရမှာဖြစ်ပါတယ်။
# တကယ်လို့ ကိုယ့်ရှာချင်တဲ့ column မှာ missing data တွေပါရင် df.col.unique(dropna=False) ဆိုပြီးရေးပေးပါ။
```

```python
# number of unique
# unique ဖြစ်တဲ့ အရေအတွက်ကိုကြည့်ချင်ရင်တော့ nunique() ကို သုံးရပါတယ်။
df.color.nunique()
# df['color'].nunique()
```

```python
# တန်ဖိုးအကြီးဆုံး values များကို nlargest ဖြင့်ကြည့်ခြင်း
df['price'].nlargest(10)
# nlargest() ထဲမှာ တန်ဖိုးမပေးထားရင် ၅ ခုကိုသာ ထုတ်ပြပါတယ်။
```

```python
# တန်ဖိုးငယ်သည်များကို nsmallest() ဖြင့်ရှာခြင်ူ
df['price'].nsmallest()
```

```python
# column မှ data frame တခုလုံး၏ အကြီးဆုံးတန်ဖိုးကို ရှာခြင်း
df.nlargest(10,['price'])
```

```python
# တခုထက်ပိုသည့် column များမှ data frame တခုလုံး၏ အကြီးဆုံးတန်ဖိုးကို ရှာခြင်း
df.nlargest(10,['price','table'])
# ဒီနည်းကိုတော့ သိပ်မသုံးသင့်ပါဘူး။ ဘာလို့လဲဆိုတော့ ပထမရေးတဲ့ column ကိုပဲမူတည်ပြီးရှာလို့ပါ။
```
