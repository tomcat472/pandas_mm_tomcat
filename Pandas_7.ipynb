```python
# set index method ဖြင့် column မှ တန်ဖိုးများအား index အဖြစ်သတ်မှတ်ခြင်း
df.set_index('cut',inplace=True)

# ဒီမှာဆိုရင် cut ဆိုတဲ့ column ကို index အဖြစ်သတ်မှတ်လိုက်တာဖြစ်ပါတယ်။
# တခုရှိတာက inplace=False ထားရင်ပြန် assign လုပ်ရမှာဖြစ်ပါတယ်။
# ကိုယ့်ရဲ့ data ကို မထိခိုက်စေချင်ဘူးဆိုရင် variable name အသစ်ပေးပြီး assign လုပ်သင့်ပါတယ်။
```

```python
df
```

```python
# index များအား sort လုပ်ခြင်း
df.sort_index(ascending=False)
# ascending=False က ပြောင်းပြန်စီတာဖြစ်ပါတယ်။
```

```python
df.sort_index(ascending=True)
# ascending=True က အတည့်စီတာဖြစ်ပါတယ်။
```

```python
# index ကို reset ပြုလုပ်ခြင်း
# ဆိုလိုတာက မူလအတိုင်း index တွေကို ပြန်ထားတာဖြစ်ပါတယ်။
df.reset_index(inplace=True)
# inplace ကို True ထားပေးရင် ထုံးစံအတိုင်း assign လုပ်စရာမလိုပါဘူး။
```

```python
df
```

```python
df.cut.value_counts()
# value count method က ကိုယ်ရချင်တဲ့ column က value တခုချင်းစီရဲ့ တန်ဖိုးစုစုပေါင်းကို ရေတွက်ပေးပါတယ်။
```

```python
# value count လုပ်ထားတဲ့ column ၏ index အား sort လုပ်ခြင်း
df.cut.value_counts().sort_index()
# ascending တွေထည့်ရင်လည်းရပါတယ်။
```

```python
# အပေါ်က sort လုပ်ထားတဲ့ data များအား bar chart ပြုလုပ်ခြင်း
df.cut.value_counts().sort_index().plot(kind='bar')
# ဒီမှာက အက္ခရာစဉ်နှင့် sort လုပ်တာဖြစ်ပါတယ်။
```

```python
# column ထဲမှ value များအား sort ပြုလုပ်ခြင်း
# ဒီမှာက numerical value များဖြင့် sort လုပ်တာဖြစ်ပါတယ်။
# ပထမဆုံးလုပ်ရမှာက ကိုယ်လုပ်မယ့် column ကိုရွေးပြီး sort_values() လုပ်ရမှာဖြစ်ပါတယ်။
# အဲ့တာပြီးမှ value count လုပ်ရမှာဖြစ်ပါတယ်။
df.cut.sort_values().value_counts()
```

```python
# အပေါ်က data အား bar chart လုပ်ခြင်း
df.cut.sort_values().value_counts().plot(kind='bar')
```
