```python
# Pandas plotting
# အရှေ့ file တွေမှာတုန်းက analysis လုပ်တာဖြစ်ပြီး ဒီကနေစပြီးတော့ visualization ဘက်ကို သွားမှာဖြစ်ပါတယ်။
# ဒီမှာသုံးသွားမှာကတော့ matplotlib နဲ့ pandas ကို ညှပ်ပြီးသုံးသွားမှာဖြစ်ပါတယ်။
# matplotlib အတွက်ကို သပ်သပ် ထပ်မံရှင်းပြသွားပါမယ်။
```

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# google drive နှင့်လင့်ချိတ်ခြင်း
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/diamonds.csv')
df.head()
```

```python
# တခုသတိထားရမှာက numeric data နဲ့ text data ကို သတိထားရမှာပါ။
# သူတို့အပေါ်မူတည်ပြီး plot လုပ်တဲ့ method အမျိုးမျိုးရှိပါတယ်။
```

```python
# line plot
df.price.head(25).plot()
# df.price.plot(kind='line)
# ဒါဆိုရင် line plot တခုထွက်ခဲ့ပါပြီ။
```

```python
# histogram
df.price.hist()
# ဒီပုံကိုကြည့်ရင်တော့ အပေါ်က ပုံနဲ့မတူဘဲ ရှင်းရှင်းလင်းလင်းမြင်ရမှာဖြစ်ပါတယ်။
```

```python
df.price.hist(bins=20,color='red',grid=False)
# bins က bar အရေအတွက် ဖြစ်ပြီး color ကတော့ အရောင်ဖြစ်ပါတယ်။
# grid True ဆိုရင် grid line ဖော်ပြတာဖြစ်ပြီး false ကျတော့ grid line off တာဖြစ်ပါတယ်။
```

```python
# bar plot
# bar plot ကို များသောအားဖြင့် text data တွေမှာသုံးလေ့ရှိတယ်။
df.cut.value_counts().plot(kind='bar',rot=0)
# rot က အောက်က label rotation လုပ်တာဖြစ်ပါတယ်။
```

```python
# horizontal bar
df.cut.value_counts().plot(kind='barh',rot=0)
```

```python
# အရောင်တွေစိတ်ကြိုက်သုံးချင်ရင်တော့ cmaps ကိုသုံးပါတယ်။
df.cut.value_counts().plot(kind='bar',rot=0,cmap='Set3')
# cmap မှာ color rand အများကြီးရှိပါတယ်။
```

```python
# column နှစ်ခုကိုသုံးပြီး အမျိုးအစားမတူသောအကြောင်းအရာများ မည်မျှရှိမည်ကို plotting လုပ်ခြင်း
df[['cut','color']].value_counts().sort_values(ascending=False).plot(kind='barh')
```

```python
# အပေါ်ကပုံကိုကြည့်လိုက်ရင် label တွေကပ်နေတာကို တွေ့ရမှာဖြစ်ပါတယ်။
# အဲ့လိုမဖြစ်ဖို့အတွက် figurer setting ကိုချိန်ရပါမယ်။
```

```python
fig,ax=plt.subplots(1,1,figsize=(6,8))
df[['cut','color']].value_counts().sort_values(ascending=False).plot(kind='barh')
# padding တွေအရမ်းများနေတဲ့အတွက်ပုံက ကြည့်လို့မလှပါဘူး။
```

```python
fig,ax=plt.subplots(1,1,figsize=(6,8))
df[['cut','color']].value_counts().sort_values(ascending=False).plot(kind='barh')
plt.tight_layout(pad=5)
```

```python
# figure style တွေကိုရချင်ရင် plt.style.available ဆိုပြီးရှာလို့ရပါတယ်။
plt.style.available
```

```python
# အဲ့ထဲကနေကိုယ်သုံးချင်တဲ့ဟာကို plt.style.use('')နဲ့သုံးလို့ရပါတယ်။
plt.style.use('seaborn-dark-palette')
```

```python
df[['price','table']].head(10).plot(kind='bar',stacked=True)
```

```python
df.hist();
# df ကို တောက်လျှောက် plotting လုပ်ရင် ရှိသမျှ data တွေအားကို plot လုပ်တယ်ဆိုတာ တွေ့ရမှာဖြစ်ပါတယ်။
# ပုံလှချင်ရင်တော့ ကိုယ့်ဘာသာ ချိန်ညှိမှ ရမှာပါ။
```

```python
# pie chart
df.color.value_counts().plot(kind='pie',autopct='%1.2f%%',shadow=True,
                             explode=(0,0,0,0.13,0,0,0),cmap= 'Set3_r');
# autopct ဆို အမျိုးအစားတခုချင်းစီအလိုက် ပါဝင်တဲ့ percentage ဖြစ်ပါတယ်။
# explode ဆိုတာ ကိုယ် hightlight လုပ်ချင်တဲ့အရာကို ဖဲ့ပြီးထုတ်ပြတာဖြစ်ပါတယ်။
```

```python
# ပုံနဲ့ပတ်သက်တဲ့ styling တွေကို matplotlib ကြမှ အသေးစိတ်ပြန်ပြောပြပါမယ်။
```

```python
# မတူညီတဲ့ data တွေကို ပေါင်းပြီးဖော်ပြခြင်း
# အဲ့လိုဖော်ပြမယ်ဆိုရင် python oop ပုံစံမျိုးနဲ့ဖော်ပြမှာဖြစ်ပါတယ်။

fig,axs=plt.subplots(2,2,figsize=(10,10))
# (2,2) ဆိုတာက obj(4) ခုထည့်မယ်လို့ဆိုလိုတာပါ။ row 0 မှာ obj 2 ခု row 1 မှာ obj 2 ခုထည့်မယ်လို့ဆိုလိုတာဖြစ်ပါတယ်။
# fig size ဆိုတာ ပုံအရွယ်အစားဖြစ်ပါတယ်။

df.cut.value_counts().plot(kind='bar',rot=0,ax=axs[0][0],title='smaple 1')


df.cut.value_counts().plot(kind='barh',rot=0,ax=axs[0][1], title='sample 2')

df.color.value_counts().plot(kind='pie',autopct='%1.2f%%',shadow=True,
                             explode=(0,0,0,0.13,0,0,0),cmap= 'Set3_r',ax=axs[1][0], title='sample 3')

df.price.hist(bins=20,color='red',grid=False,ax=axs[1][1])

plt.suptitle('Samples',fontsize=20)
# suptitle က ခေါင်းစဉ်ကြီးဖြစ်ပါတယ်။

plt.tight_layout(pad=3)
# tight layout က စာသားတွေ တခုနဲ့တခု မထပ်ဖို့အတွက်သုံးပါတယ်။
# ပုံအလိုက် x,y ဝင်ရိုးမှာ စာလည်းထည့်လို့ရပါတယ်။
# နောက်တခုကို နောက်တနည်းနဲ့ဆက်လုပ်ရအောင်။
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Samples',fontsize=20)


plt.subplot(2,2,1)
plt.title('sample 1')
df.cut.value_counts().plot(kind='bar',rot=0)

plt.subplot(2,2,2)
plt.title('sample 2')
df.cut.value_counts().plot(kind='barh',rot=0)

plt.subplot(2,2,3)
plt.title('sample 3')
df.color.value_counts().plot(kind='pie',autopct='%1.2f%%',shadow=True,
                             explode=(0,0,0,0.13,0,0,0),cmap= 'Set3_r')

plt.subplot(2,2,4)
plt.title('sample 4')
df.price.hist(bins=20,color='red',grid=False)

plt.tight_layout(pad=3);
```
