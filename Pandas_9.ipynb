```python
df=pd.read_csv('/content/drive/MyDrive/diamonds.csv')
df.head()
```

```python
# မိမိရချင်သော data များကိုထုတ်ခြင်း
## ပထမဆုံးအနေဖြင့် string များကို ထုတ်တဲ့ဟာအရင်လုပ်ပါမယ်။
df.cut.str.contains('Good')
# ဒီမှာဆိုရင် Good စာလုံးပေါင်းပါသမျှဟာအားလုံးကို ထုတ်ပေးတာဖြစ်ပါတယ်။
# data frame အနေဖြင့်မြင့်ချင်ရင်သော nested လုပ်မှ မြင်ရမှာဖြစ်ပါတယ်။
```

```python
df[df.cut.str.contains('Good')]
```

```python
# အပေါ်ကစစ်ထုတ်တာရဲ့ ဆန့်ကျင်ဘက်ကို ရချင်ရင် ~ sign သုံးပေးရပါတယ်။ ၎င်းသင်္ကေတသည် negative ဖြစ်ပါတယ်။
df[~(df.cut.str.contains('Good'))]
```

```python
# အပေါ်က နည်းကိုသုံးမယ်ဆိုရင် ကိုယ်စာမယ့်စာလုံးအကြီးအသေးကို သတိထားပြီးလုပ်ရပါတယ်။
# အဲ့လိုပြဿနာများရှိသည့်အတွက် re library နှင့်တွဲသုံးမှ အဆင်ပြေပါတယ်။
```

```python
import re
df.loc[df['cut'].str.contains('good[a-z]*',flags=re.I,regex=True)]
# ဒီနည်းနဲ့ ရှာရင်ကျတော့ good ပါတဲ့ row အားလုံးကို ထုတ်ပေးပါတယ်။
```

```python
df.loc[df['cut'].str.contains('^good[a-z]*',flags=re.I,regex=True)]
# ဒီနည်းနှင့်ရှာရင်ကျတော့ good သီးသန့်ပဲထုတ်ပေးပါတယ်။
```

```python
# မိမိရချင်သော စာသားများအား isin method ဖြင့်ထုတ်ခြင်း
A=['Good','Very Good']
df[df.cut.isin(A)]
# df[df.cut.isin(['Good','Very Good'])]
```

```python
# ကြားတန်ဖိုးများကိုရှာခြင်း
# 100 <= x <= 200 
# ဒီမှာဆိုရင် x က ၁၀၀ ကနေ ၂၀၀ ကြားလို့အဓိပ္ပါယ်ရပါတယ်။
# အဲ့ဒီလိုပုံစံမျိုးကိုရှာမှာဖြစ်ပါတယ်။
```

```python
# price colunm က ၄၀၀ နဲ့ ၅၀၀ ကြားကို ရှာခြင်း
df.loc[(df.price>=400) & (df.price<=500)]
# ဒီနည်းကို အရှေ့မှာ ရှင်းပြပြီးကြောင့်မို့ ဆက်မရှင်းပြတော့ပါဘူး။
```

```python
# အပေါ်ကနည်းထက် ဒီနည်းက မှတ်ရတာ ပိုလွယ်ပါတယ်။
df[df.price.between(400,500)]
```

```python
# 400 နှင့် 500 ကြားရှိတန်ဘိုးများအား ရေတွက်ခြင်း
df[df.price.between(400,500)].value_counts()
```
