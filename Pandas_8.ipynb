```python
# removing columns
# ကိုယ်မရချင်တဲ့ columns တွေကို ဖယ်တဲ့အခါမှာ မူရင်း data ကို မထိခိုက်ဖို့အတွက် variable အသစ်တခုလုပ်သင့်ပါတယ်။
df_new=df.drop(columns=['x'])
df_new
```

```python
# တခုထက်ပိုတဲ့ columns များကို remove လုပ်ခြင်း
df_new=df_new.drop(columns=['y','z'])
df_new
```

```python
# variable assign မလုပ်ဘဲနဲ့ သုံးမယ်ဆိုရင် inplace=True ကိုရွေးပေးရမှာဖြစ်ပါတယ်။
# နောက်ထပ် remove လုပ်တဲ့နည်းကတော့ data ကို define လုပ်တဲ့အချိန်မှာ ကိုယ်သုံးချင်တဲ့ column ကိုရွေးပေးလို့ရပါတယ်။
# ဒါပေမယ့်ဒီနည်းကိုလည်းမသုံးသင့်ပါဘူး။
```

```python
# နောက်တနည်းကတော့ columns တွေစုပြီး data frame အသစ်တခုလုပ်တဲ့နည်းဖြစ်ပါတယ်။
another_df=df[['cut','color','price']]
another_df.head()
```

```python
# အပေါ်က drop method က columns တွေကိုသာမက rows တွေကိုပါ drop လုပ်လို့ရပါတယ်။
# row တွေကို drop လုပ်မယ်ဆိုရင် ကိုယ်မလိုချင်တဲ့ data တွေရှိတဲ့ column ကို index အဖြစ် သတ်မှတ်ပေးသင့်ပါတယ်။
another_df.set_index('cut',inplace=True)
```

```python
another_df.sort_index()
```

```python
# ဒီမှာဆိုရင် fair ပါတဲ့ row တွေကို remove လုပ်ပါမယ်။
# inplace=True လုပ်ပေးရပါမယ်။
# axis=0 ဆိုတာက row ကိုပြောတာဖြစ်ပါတယ်။
another_df.drop(labels='Fair',axis=0,inplace=True)
```

```python
# တခုထက်ပိုတဲ့ row တွေကို remove လုပ်ခြင်း
another_df.drop(labels=['Good','Very Good'],axis=0,inplace=True)
```

```python
# ဘယ်လိုသွားမှာလဲဆိုတော့ dictionary ရဲ့သဘောတရားကို သုံးပြီး နာမည်ပြောင်းမှာဖြစ်ပါတယ်။
```

```python
# Dictionary မှာ ရဲ့ data types မှာ  key နဲ့ value ဆိုပြီး အတွဲလေးတွေပါတယ်။
# java script က json နဲ့သဘောသဘာဝချင်း ဆင်တူပါတယ်။
d={
    'name':'Tom',
    'age':25,
    'town':'Ygn'
}
print(d)
print(type(d))
```

```python
# အရှေ့က name,age,town တို့က key ဖြစ်ပြီး string ဖြင့်ရေးပါသည်။
for key in d.keys():
  print(key)
```

```python
for value in d.values():
  print(value)
```

```python
# key နဲ့ value ကို အတူတူ ထုတ်ချင်ရင် .items() ကိုသုံးရပါတယ်။
d.items()
```
